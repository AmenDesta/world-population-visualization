This notebook shows how global population has grown from 1950 to 2025 using simple charts.  
It uses Python libraries like `pandas` and `matplotlib` to visualize the data.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Year": [1950, 1970, 1990, 2010, 2020, 2025],
    "Population (Billions)": [2.5, 3.7, 5.3, 6.9, 7.8, 8.1]
}
df = pd.DataFrame(data)
```

## Line Chart: Population Growth Over Time

This chart shows how the global population has steadily increased over the decades.


```python
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Population (Billions)"], marker="o", color="navy")
plt.title("World Population Growth (1950‚Äì2025)")
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.grid(True)
plt.show()
```

## Bar Chart: Year-by-Year Comparison

This chart compares population size across selected years.


```python
plt.figure(figsize=(8,5))
plt.bar(df["Year"], df["Population (Billions)"], color="teal")
plt.title("World Population by Year")
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.show()
```

## Insights

- The global population more than tripled between 1950 and 2025.  
- The steepest growth occurred between 1970 and 2010.  
- By 2025, the world is projected to reach over 8 billion people.


```python
growth = df["Population (Billions)"].iloc[-1] - df["Population (Billions)"].iloc[0]
print(f"Global population grew by {growth:.1f} billion between 1950 and 2025.")
```
